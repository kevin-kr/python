```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np 
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')

sample_size = 50
x = np.random.normal(-5, 5, sample_size)
y = np.random.normal(-5, 5, sample_size)
z = np.random.normal(-5, 5, sample_size)

ax.plot(x, y, z, alpha=0.5, marker='o')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')

x = np.cumsum(np.random.normal(0, 5, 1000))
y = np.cumsum(np.random.normal(0, 5, 1000))
z = np.cumsum(np.random.normal(0, 5, 1000))

ax.scatter(x, y, z, c = z, s= 20, alpha=0.5, cmap=plt.cm.Greens)
plt.show()
```
